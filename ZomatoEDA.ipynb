```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv("/content/zomato.csv",encoding="latin-1")
df.head()
```

```python
df_country=pd.read_excel("/content/Country-Code.xlsx")
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(5)
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",palette=["white","red","orange","yellow","green","green"],data=ratings)
```

```python
# Count plot
sns.countplot(x="Rating color",data=ratings,palette=["white","red","orange","yellow","green","green"])
```

```python
# Find the countries name that has given 0 rating
final_df[final_df["Rating text"]=="None"].groupby("Country").size().reset_index()
```

```python
final_df.groupby(['Aggregate rating','Country']).size().reset_index()
```

```python
# Find which curreny used in which Country
country_currency=final_df.groupby(['Country','Currency']).size().reset_index()
country_currency
```

```python
# Find which countries has online deliverys
final_df[final_df["Has Online delivery"]=="Yes"].groupby("Country").size()
```
